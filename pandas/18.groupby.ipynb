```python
# 등급별 가격의 합(~별)
# DataFrameGroupby (집계관련함수:sum,count,max,min,agg,mean)
g = df.groupby('grade')
g.sum()[['price']] # 집계가능한 (숫자)모든컬럼이 나옴, 데이터 프레임으로 결과를 볼려면 복수넣는거처럼[]을넣고 그안에 1개만 넣는다
```

```python
g = df.groupby('grade')
g.mean()[['price']]
```

```python
g = df.groupby('grade')
g.max()[['price']]
```

```python
g = df.groupby('grade')
g.min()[['price']]
```

```python
g = df.groupby('grade')
g.count()
```

```python
g = df.groupby('grade')
mydf = g.agg(['sum','mean'])['price']
```

```python
mydf.columns = ['가격의 총합','가격의평균']
```
