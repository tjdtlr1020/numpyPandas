# 창을 분할하는 방법들

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(2,2,1) # 2행2열로 나눳을때 1번쨰 표에다가 그려라
plt.plot(x,y1)
plt.subplot(2,2,2) # 2행2열로 나눳을때 2번쨰 표에다가 그려라
plt.plot(x,y2)
plt.subplot(2,2,3) # 2행2열로 나눳을때 3번쨰 표에다가 그려라
plt.plot(x,y2)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(1,3,1) # 2행2열로 나눳을때 1번쨰 표에다가 그려라
plt.plot(x,y1)
plt.subplot(1,3,2) # 2행2열로 나눳을때 2번쨰 표에다가 그려라
plt.plot(x,y2)
plt.subplot(1,3,3) # 2행2열로 나눳을때 3번쨰 표에다가 그려라
plt.plot(x,y2)
plt.show()
```

# 범례 표시와 그리드표시

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.xlabel('x축라벨')
plt.ylabel('y축라벨')
plt.grid(True)
plt.title('타이틀')
plt.plot(x,y1,'bo-',x,y2,'rs--',x,y3,'yp-.')
plt.legend(['국어','영어','수학'])
plt.show()
# fmt[marker][line][color]ow()
```

# text넣는거 포문돌리기

```python
xname = ['홍길동','이순신','임꺽정','김철수','이황']
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적')
plt.grid(True)
plt.xticks(x,xname,rotation = 45) # x라는 값을 xname으로 매핑 좌표값으로 표시해야 텍스트 추가시 편하게할수있다

plt.plot(x,y1,'ro--',alpha = 0.5) # alpha = 선의 투명도 설정 0~1
plt.text(1,10,'10') # 1,10 에 '10' 이라는 텍스트 추가
for n1,n2 in zip(x,y1):
    plt.text(n1,n2,f'{n2}점')
plt.legend(['국어'])
plt.show()
```

# 하나의창에 2개의 그래프 y축이 다르게

```python
fig,ax1 = plt.subplots() # s가붙음
ax1.plot(x,y1)
ax1.set_xlabel('x축')     #axes객체의 x축이름은xlabel이 아닌 set_xlabel
ax1.set_ylabel('y1축',color = 'blue') 
ax2 = ax1.twinx()        # y축을 오른쪽 기준으로 잡은 그래프 반환
ax2.set_ylabel('y2축',color = 'red',rotation = 90)
ax2.plot(x,y2,'r-')
plt.show()
```
