```python
def fn(v):
#     print('='*10)
#     print(v)
#     print('='*10)
    if v.mean()>=70:
        return '합격'
    else:
        return '불합격'
#     return v.mean()
```

```python
# gradeDF = df.pivot_table(index = '학년',values = ['중간'],aggfunc = ['mean',fn])
gradeDF = df.pivot_table(index = '학년',values = ['중간'],aggfunc = fn)
gradeDF
```

```python
gradeDF = df.pivot_table(index = '학년',values = ['중간'],aggfunc = lambda v:'합격' if v.mean()>=70 else '불합격')
gradeDF
```
